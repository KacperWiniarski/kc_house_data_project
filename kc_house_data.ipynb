```python
# zdefinuj zmienne X

#X = df[['bathrooms', 'sqft_living', 'grade', 'sqft_above', 'sqft_living15']]
#X = df[['sqft_living', 'grade']]
X = df.drop(['date','price'], axis=1)
```

```python
# zdefiniuj zmienną y
y = df['price']
```

```python
# podziel dane na zbiór treningowy i testowy
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2)
```

# Regresja liniowa

```python
# wytrenuj model regresji liniowej na zbiorze treningowym
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# zrób predykcje na zbiorze testowym
y_pred = model.predict(X_test)
```

```python
# oblicz współczynnik r2
r2_score(y_test, y_pred)
```

```python
# oblicz metrykę MSE
mean_absolute_error(y_test, y_pred)
```

```python
# wyświetl współczynniki dla zmiennych
model.coef_
```

```python
# wyświetl współczynnik dla wyrazu wolnego
model.intercept_
```

# Drzewo decyzyjne

```python
# wytrenuj model na zbiorze treningowym
model=DecisionTreeRegressor(max_depth=8)
model.fit(X_train, y_train)
```

```python
# zrób predykcje na zbiorze testowym
y_pred_tree = model.predict(X_test)
```

```python
# oblicz współczynnik r2
r2_score(y_test, y_pred_tree)
```

```python
# oblicz metrykę MSE
mean_absolute_error(y_test, y_pred_tree)
```

```python
# narysuj model drzewa
fig, ax=plt.subplots(figsize=(30,24))
ax=tree.plot_tree(model)
plt.show()
```
